```python
from pymongo import MongoClient
from pprint import pprint
import json

client = MongoClient('mongodb://localhost:27017')
db = client.spire
status = db.command('serverStatus')
runs = db.runs

result = runs.aggregate([
    {
        '$project': {
            'event.event_choices': 1
        }
    }, {
        '$unwind': {
            'path': '$event.event_choices'
        }
    }, {
        '$project': {
            'arrayofkeyvalue': {
                '$objectToArray': '$event.event_choices'
            }
        }
    }, {
        '$unwind': {
            'path': '$arrayofkeyvalue'
        }
    }, {
        '$group': {
            '_id': '$arrayofkeyvalue.k'
        }
    }
])
[item['_id'] for item in result]
```

```python
results = runs.aggregate([
    {
        "$match": {
            "event.character_chosen": "IRONCLAD",
            "event.ascension_level": {"$gte": 15},
            "event.floor_reached": {"$gte": 40},
            "event.event_choices.cards_transformed": {"$exists": True}
        }
    },
    {
        "$sample": {'size': 1}
    },
])
result = results.next()
pprint(result['event']['event_choices'])
```
